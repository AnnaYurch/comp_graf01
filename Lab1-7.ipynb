10. Как определить, есть ли в данном двумерном массиве пустые столбцы?

````python
def find_empty_columns(arr, empty_value=0):
    if np.isnan(empty_value):
        empty_mask = np.all(np.isnan(arr), axis=0)
    else:
        empty_mask = np.all(arr == empty_value, axis=0)
    
    empty_indices = np.where(empty_mask)[0]
    return empty_mask, empty_indices

arr = np.array([[1, 0, 0, np.nan],
                [4, 0, 0, np.nan],
                [7, 0, 0, np.nan]])

empty_mask, empty_indices = find_empty_columns(arr, empty_value=0)
print("Пустые столбцы (нули):", empty_indices)

empty_mask, empty_indices = find_empty_columns(arr, empty_value=np.nan)
print("Пустые столбцы (NaN):", empty_indices)
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
from scipy import stats

df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure (s/d)'].str.split('/', expand=True).astype(float)

stress_high = df[df['Stress Levels'] >= df['Stress Levels'].quantile(0.75)]
stress_low = df[df['Stress Levels'] <= df['Stress Levels'].quantile(0.25)]

systolic_high = stress_high['Systolic_BP'].mean()
systolic_low = stress_low['Systolic_BP'].mean()
diastolic_high = stress_high['Diastolic_BP'].mean()
diastolic_low = stress_low['Diastolic_BP'].mean()

t_systolic, p_systolic = stats.ttest_ind(stress_high['Systolic_BP'], stress_low['Systolic_BP'], nan_policy='omit')
t_diastolic, p_diastolic = stats.ttest_ind(stress_high['Diastolic_BP'], stress_low['Diastolic_BP'], nan_policy='omit')

print(f"СИСТОЛИЧЕСКОЕ ДАВЛЕНИЕ:")
print(f"Высокий стресс: {systolic_high:.1f} mmHg")
print(f"Низкий стресс: {systolic_low:.1f} mmHg") 
print(f"Разница: {systolic_high - systolic_low:.1f} mmHg")
print(f"p-value: {p_systolic:.4f}")
print(f"Статистически значимо: {'Да' if p_systolic < 0.05 else 'Нет'}")
print()
print(f"ДИАСТОЛИЧЕСКОЕ ДАВЛЕНИЕ:")
print(f"Высокий стресс: {diastolic_high:.1f} mmHg")
print(f"Низкий стресс: {diastolic_low:.1f} mmHg")
print(f"Разница: {diastolic_high - diastolic_low:.1f} mmHg")
print(f"p-value: {p_diastolic:.4f}")
print(f"Статистически значимо: {'Да' if p_diastolic < 0.05 else 'Нет'}")
````

6. Влияет ли физическая активность на вес в возрасте более 50 лет?

````python
older_adults = df[df['Age (years)'] > 50]

activity_levels = older_adults['Physical Activity Level'].unique()
print("Доступные уровни активности:", activity_levels)

results = {}

for level in activity_levels:
    weight_data = older_adults[older_adults['Physical Activity Level'] == level]['Weight (kg)']
    results[level] = {
        'mean_weight': weight_data.mean(),
        'std_weight': weight_data.std(),
    }
    print(f"\n{level} активность:")
    print(f"  Средний вес: {results[level]['mean_weight']:.1f} кг")

groups = [older_adults[older_adults['Physical Activity Level'] == level]['Weight (kg)'] 
          for level in activity_levels]

groups = [group for group in groups if len(group) > 0]

if len(groups) >= 2:
    f_stat, p_value = stats.f_oneway(*groups)
    
    print(f"\n=== СТАТИСТИЧЕСКИЙ АНАЛИЗ ===")
    print(f"p-value: {p_value:.4f}")
    
    if p_value < 0.05:
        print("Физическая активность ЗНАЧИМО влияет на вес у людей старше 50 лет")
    else:
        print("Физическая активность НЕ оказывает значимого влияния на вес у людей старше 50 лет")
else:
    print("Недостаточно данных")
````

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
high_income = df[df['Income Level'] == 'High']
low_income = df[df['Income Level'] == 'Low']

print("=== СРАВНЕНИЕ УРОВНЯ ФИЗИЧЕСКОЙ АКТИВНОСТИ ===")

def activity_to_number(level):
    if level == 'Low':
        return 1
    elif level == 'Moderate':
        return 2
    elif level == 'High':
        return 3
    else:
        return 0  

high_income_numeric = high_income['Physical Activity Level'].apply(activity_to_number)
low_income_numeric = low_income['Physical Activity Level'].apply(activity_to_number)

high_income_mean = high_income_numeric.mean()
low_income_mean = low_income_numeric.mean()

print(f"\nРАСПРЕДЕЛЕНИЕ ПО УРОВНЯМ АКТИВНОСТИ:")

for level in ['High', 'Moderate', 'Low']:
    high_count = (high_income['Physical Activity Level'] == level).sum()
    low_count = (low_income['Physical Activity Level'] == level).sum()
    high_percent = (high_count / len(high_income)) * 100
    low_percent = (low_count / len(low_income)) * 100
    
    print(f"{level}:")
    print(f"  Высокий доход: {high_count} чел. ({high_percent:.1f}%)")
    print(f"  Низкий доход: {low_count} чел. ({low_percent:.1f}%)")

print(f"\n=== ОТВЕТ ===")
if high_income_mean > low_income_mean:
    print(f"Люди с высоким доходом имеют БОЛЬШИЙ уровень физической активности")
else:
    print(f"Люди с высоким доходом имеют МЕНЬШИЙ уровень физической активности")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data.csv')

# Создаем возрастные группы
df['Age Group'] = pd.cut(df['Age (years)'], 
                        bins=[0, 30, 60, 100], 
                        labels=['Молодые', 'Средние', 'Пожилые'])

plt.figure(figsize=(15, 5))

for i, age_group in enumerate(df['Age Group'].unique()):
    plt.subplot(1, 3, i+1)
    group_data = df[df['Age Group'] == age_group].dropna()
    
    # Точечный график
    plt.scatter(group_data['Vision Sharpness'], group_data['Hearing Ability (dB)'],
                alpha=0.6, s=30)
    
    # Линия тренда
    if len(group_data) > 1:
        z = np.polyfit(group_data['Vision Sharpness'], group_data['Hearing Ability (dB)'], 1)
        p = np.poly1d(z)
        plt.plot(group_data['Vision Sharpness'], p(group_data['Vision Sharpness']), "red", linewidth=2)
    
    plt.title(f'{age_group} ({len(group_data)} чел.)')
    plt.xlabel('Острота зрения')
    plt.ylabel('Слух (dB)')

plt.tight_layout()
plt.show()

# Анализ корреляции
print("=== АНАЛИЗ СВЯЗИ ЗРЕНИЯ И СЛУХА ===")
for age_group in df['Age Group'].unique():
    group_data = df[df['Age Group'] == age_group].dropna()
    corr = group_data['Vision Sharpness'].corr(group_data['Hearing Ability (dB)'])
    print(f"{age_group}: корреляция = {corr:.3f}")

print("\n=== ОТВЕТ НА ВОПРОС ===")
print("Связь между зрением и слухом СЛАБАЯ (корреляция близка к 0)")
print("Зависимость: практически отсутствует")
print("Почему: зрение и слух - независимые сенсорные системы")
print("Оба ухудшаются с возрастом, но не влияют друг на друга напрямую")
````
